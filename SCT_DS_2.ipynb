**Import Libraries**

```python
import pandas as pd
import matplotlib.pyplot as plt
```

**Assign the CSV file to a variable**

```python
df= pd.read_csv("/Users/macbook/Downloads/API_SP.POP.TOTL_DS2_en_csv_v2_34/ph_dengue_cases2016-2020.csv")
df.head()
```

# Exploratory Data Analysis (EDA)
Performing EDA to understand the dataset.

```python
# Display the first few rows of the dataset
print("First 5 rows of the dataset:")
print(df.head())

# Check the shape of the dataset
print("\nShape of the dataset:")
print(df.shape)

# Check for missing values
print("\nMissing values in each column:")
print(df.isnull().sum())

# Check the data types of each column
print("\nData types of each column:")
print(df.dtypes)

# Summary statistics of numerical columns
print("\nSummary statistics of numerical columns:")
print(df.describe())

# Check the unique values in the 'Region' column
print("\nUnique regions in the dataset:")
print(df['Region'].unique())

# Check the distribution of dengue cases
print("\nDistribution of Dengue Cases:")
print(df['Dengue_Cases'].describe())

# Check the distribution of dengue deaths
print("\nDistribution of Dengue Deaths:")
print(df['Dengue_Deaths'].describe())
```

**Visualization**

```python

# Check for duplicates
print("Number of duplicate rows:", df.duplicated().sum())

# Remove duplicates if any
df = df.drop_duplicates()
print("Shape of the dataset after removing duplicates:", df.shape)

# Convert 'Month' to a categorical variable with proper ordering
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 
               'July', 'August', 'September', 'October', 'November', 'December']
df['Month'] = pd.Categorical(df['Month'], categories=month_order, ordered=True)

# Explore relationships between variables
# Plot Dengue Cases over the years
plt.figure(figsize=(10, 6))
df.groupby('Year')['Dengue_Cases'].sum().plot(kind='bar', color='skyblue')
plt.title('Total Dengue Cases per Year')
plt.xlabel('Year')
plt.ylabel('Total Dengue Cases')
plt.show()

# Plot Dengue Cases by Region
plt.figure(figsize=(12, 6))
df.groupby('Region')['Dengue_Cases'].sum().sort_values(ascending=False).plot(kind='bar', color='orange')
plt.title('Total Dengue Cases by Region')
plt.xlabel('Region')
plt.ylabel('Total Dengue Cases')
plt.show()

# Analyze trends in Dengue Cases over months
plt.figure(figsize=(12, 6))
df.groupby('Month')['Dengue_Cases'].sum().reindex(month_order).plot(kind='line', marker='o', color='green')
plt.title('Monthly Trend of Dengue Cases')
plt.xlabel('Month')
plt.ylabel('Total Dengue Cases')
plt.grid()
plt.show()

# Analyze relationship between Dengue Cases and Dengue Deaths
plt.figure(figsize=(8, 6))
plt.scatter(df['Dengue_Cases'], df['Dengue_Deaths'], alpha=0.5, color='purple')
plt.title('Relationship between Dengue Cases and Dengue Deaths')
plt.xlabel('Dengue Cases')
plt.ylabel('Dengue Deaths')
plt.grid()
plt.show()
```
